```python
import re
import pandas as pd
from pprint import pprint
import nltk
from konlpy.tag import Okt
okt = Okt()
```

```python
df = pd.read_csv('./data/음식점_리뷰_결과_정리.csv', index_col=False)
df = df[df['total']>=50]
df = df[df['percent']<=85]
display(df)
title_list = df['title']
title_list = list(title_list)
```

```python
bad_emotion = {'맛':['짜다','냄새','노맛','느끼하다','밋밋','비리다','싱겁다',
                     '맛없다','조미료','남기다','밍밍','무맛','약하다','쎄다','밋밋하다',
                     '오래되다','달라지다','짠맛','쓴맛','단맛','변하다'],
               '서비스':['불친절','불친절하다','최악','기다림','누락','잡담','핑계','비위생적',
                         '싸가지','강요','실망','더럽다','아깝다','시끄럽다','좁다'],
               '양':['부족하다','적다'],
               '가격':['비싸다','창렬함','창렬']}
```

```python
print(okt.pos('덜렁',stem=True))
print(okt.pos('짜고 짜서 짜다 짰다 짰지만 짰어요',stem=True))
print(okt.pos('셔요 시다 신맛 느끼한 비려서 비린내 비렸다 비렸지만 창렬 창렬이네요',stem=True))
print(okt.pos('셔요 시다 신맛 느끼한 비려서 비린내 비렸다 비렸지만 창렬 창렬이네요',norm=True))
```

```python
result = []

for t in title_list:
    taste = []
    service = []
    quantity = []
    price = []
    review = []
    
    for i in df[(df['title'] == t) & (df['predict'] == 0)]['review']:
        okt_pos = [t[0] for t in okt.pos(i)]
        okt_pos_stem = [t[0] for t in okt.pos(i,stem=True)]

        index_t = []
        index_s = []
        index_q = []
        index_p = []
        
        for j in okt_pos_stem:
            for k in bad_emotion['맛']:
                if k == j:
                    taste.append(k)
                    index_t.append(okt_pos_stem.index(j))

            for k in bad_emotion['서비스']:
                if k == j:
                    service.append(k)
                    index_s.append(okt_pos_stem.index(j))

            for k in bad_emotion['양']:
                if k == j:
                    quantity.append(k)
                    index_q.append(okt_pos_stem.index(j))

            for k in bad_emotion['가격']:
                if k == j:
                    price.append(k)
                    index_p.append(okt_pos_stem.index(j))
        
        for index in index_t:
            if re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i):
                review.append(''.join(re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i)))
                
        for index in index_s:
            if re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i):
                review.append(''.join(re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i)))
                
        for index in index_q:
            if re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i):
                review.append(''.join(re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i)))
                
        for index in index_p:
            if re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i):
                review.append(''.join(re.findall('[가-힣]*\s*[가-힣]*\s*' + okt_pos[index] + '\s*[가-힣]*\s*[가-힣]*',i)))
                
    result.append([t,list(set(taste)),list(set(service)),list(set(quantity)),list(set(price)),list(set(review))])
```

```python
for r in result:
    print('상호명 :',r[0])
    print('맛 :',r[1])
    print('서비스 :',r[2])
    print('양 :',r[3])
    print('가격 :',r[4])
    print('리뷰내용 :',r[5])
    print()
```
